```julia
using TaylorModels, IntervalArithmetic
```

```julia
using Plots; gr()
```

```julia
t = taylor_var(2, 0, -0.5..0.5)
```

```julia
f = exp(t)
```

```julia
plot(f, label="Taylor model")
plot!(exp, label="exact", lw=2)
plot!(t->mid(f.p(t)), lw=2)
```

```julia
using Interact
```

```julia
@manipulate for n in slider(1:20, value=1)
    t = taylor_var(n, 0, -3..3)

    f = sin(t)

    plot(sin(t), ylim=(-3,3))
    plot!(t.I.lo:0.01:t.I.hi, t->sin(t), lw=2)
    plot!(t.I.lo:0.01:t.I.hi, t->mid((f.p)(t)), lw=2)
end

# plot!(exp(sin(t)))
# plot!(-0.5:0.01:0.5, t->exp(sin(t)), lw=2)
```
